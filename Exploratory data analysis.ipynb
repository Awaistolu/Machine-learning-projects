```python
# Multivariate analysis

class Multivariate_EDA:
    
    def __init__(self,col1,col2):
        self.col1 =  col1
        self.col2 = col2
        
    def scatter_plot(self):
        plt.figure(figsize=(12,6))
        sns.scatterplot(x=self.col1,y=self.col2)
        plt.show()
        
    def bar_plot(self):
        plt.figure(figsize=(12,6))
        sns.barplot(x=self.col1,y=self.col2)
        plt.show()
    
    def box_plot(self):
        plt.figure(figsize=(12,6))
        sns.boxplot(x=self.col1,y=self.col2)
        plt.show()
    
    
    def hist_plot(self):
        plt.figure(figsize=(12,6))
        sns.histplot(x=self.col1,y=self.col2)
        plt.show()
        
    def line_plot(self):
        plt.figure(figsize=(12,6))
        sns.lineplot(x=self.col1,y=self.col2)
        plt.show()
        
    def heatmap_plot(self):
        plt.figure(figsize=(12,6))
        sns.heatmap(pd.crosstab(self.col1,self.col2),annot=True)
        plt.show()
```

```python
# Univariate analysis

class Univariate_EDA:
    
    def __init__(self,col):
        self.col = col
        
        
    def count_plot(self):
        plt.figure(figsize=(12,6))
        sns.countplot(self.col)
        plt.show()
        
    def box_plot(self):
        plt.figure(figsize=(12,6))
        sns.boxplot(self.col)
        plt.show()
        
    def pie_plot(self):
        plt.figure(figsize=(12,6))
        val = pd.DataFrame(self.col).value_counts()
        plt.pie(val,labels=self.col.unique(),autopct='%2.f')
        plt.show()
    
       
    def dist_plot(self):
        plt.figure(figsize=(12,6))
        sns.distplot(self.col)
        plt.show()
```

```python
# check dataset info 

class EDA:
    def __init__(self,df):
        self.df = df
        
    def info(self):
#         first five rows of dataset
        print('\n DataFrame First Five Rows\n'+'--'*50 +f'\n{self.df.head()}\n'+'--'*50)
    
#     number of rows of dataset
         
        print('\n Number of rows\n'+'--'*50 +f'\n{self.df.shape[0]}\n'+'--'*50)
        
#         number of columns of dataset

        print('\n Number of rows\n'+'--'*50 +f'\n{self.df.shape[1]}\n'+'--'*50)
    
#     Dataset Summary

        print('\n Data Summary\n'+'--'*50 +f'\n{self.df.describe()}\n'+'--'*50)
    
#     find NaN values

        print('\n Find NaN Values \n'+'--'*50 +f'\n{self.df.isnull().sum()}\n'+'--'*50)
    
#     check dtype

        print('\n Check Datatype \n'+'--'*50 +f'\n{self.df.dtypes}\n'+'--'*50)
    
#     check correlation

        print('\n Check Correlation \n'+'--'*50 +f'\n{self.df.corr()}\n'+'--'*50)
    
```
